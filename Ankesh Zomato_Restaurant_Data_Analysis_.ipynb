```python

# üì¶ Step 1: Import Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# üìÇ Step 2: Load the Dataset
file_path = "/content/zomato.csv"
df = pd.read_csv(file_path, encoding="ISO-8859-1")

# üîç Step 3: View Basic Info
print("\nüìå First 5 Rows:")
print(df.head())

print("\nüìå Dataset Info:")
print(df.info())

print("\nüìå Missing Values:")
print(df.isnull().sum())

print("\nüìå Column Names:")
print(df.columns.tolist())
```

#What This Does:
‚úÖ Removes unused or irrelevant columns

‚úÖ Drops duplicate entries

‚úÖ Handles missing and incorrect values

‚úÖ Converts important columns like Votes and Price range to numeric

‚úÖ Final clean shape is shown

```python
# üì¶ Step 1: Import Libraries
import pandas as pd

# üìÇ Step 2: Load the Dataset
file_path = "/content/zomato.csv"
df = pd.read_csv(file_path, encoding="ISO-8859-1")

# üîç Step 3: Show shape and basic info
print("üî¢ Original Data Shape:", df.shape)
print("\nüîç Checking missing values:")
print(df.isnull().sum())

# üßπ Step 4: Drop columns with too many missing values (if any)
threshold = 0.5 * len(df)
df = df.dropna(thresh=threshold, axis=1)

# üßπ Step 5: Drop irrelevant columns
columns_to_drop = ['Restaurant ID', 'Address', 'Locality Verbose', 'Switch to order menu']
df.drop(columns=columns_to_drop, axis=1, inplace=True)

# üßπ Step 6: Drop duplicate rows
df.drop_duplicates(inplace=True)

# üßπ Step 7: Drop rows with missing values in key columns
df.dropna(subset=['Cuisines', 'Aggregate rating', 'Votes'], inplace=True)

# üßπ Step 8: Convert data types
df['Votes'] = pd.to_numeric(df['Votes'], errors='coerce')
df['Price range'] = pd.to_numeric(df['Price range'], errors='coerce')

# ‚úÖ Step 9: Reset index
df.reset_index(drop=True, inplace=True)

# üìä Step 10: Show cleaned summary
print("\n‚úÖ Cleaned Data Shape:", df.shape)
print("\nüßº Cleaned Data Preview:")
print(df.head())

print("\nüìâ Remaining Null Values:")
print(df.isnull().sum())
```

#Visual Insights Program

#1. Rating Distribution

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot rating distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['Aggregate rating'], bins=20, kde=True, color='skyblue')
plt.title("Distribution of Restaurant Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Restaurants")
plt.grid(True)
plt.show()
```

#2. Most Popular Cuisines

```python
# Top 10 cuisines
top_cuisines = df['Cuisines'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_cuisines.values, y=top_cuisines.index, palette='Set2')
plt.title("Top 10 Most Popular Cuisines")
plt.xlabel("Number of Restaurants")
plt.ylabel("Cuisines")
plt.show()
```

#3. Restaurants per City

```python
# Top 10 cities with most restaurants
top_cities = df['City'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_cities.values, y=top_cities.index, palette='pastel')
plt.title("Top 10 Cities by Restaurant Count")
plt.xlabel("Number of Restaurants")
plt.ylabel("City")
plt.show()
```

#  4. Price Range vs Average Rating

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='Price range', y='Aggregate rating', data=df, palette='coolwarm')
plt.title("Price Range vs Aggregate Rating")
plt.xlabel("Price Range (1=Cheap, 4=Expensive)")
plt.ylabel("Aggregate Rating")
plt.show()
```

#5. Online Delivery vs Rating

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='Has Online delivery', y='Aggregate rating', data=df, palette='Set3')
plt.title("Online Delivery vs Aggregate Rating")
plt.xlabel("Has Online Delivery")
plt.ylabel("Rating")
plt.show()
```

#Top 5 Restaurants Offering Online Delivery
**"The top 5 restaurants offering online delivery have ratings above 4.4 and serve cuisines like Japanese, Filipino, and Sushi. Most are located in Mandaluyong and Makati City, indicating that customers value both convenience and qualit**

```python
online_delivery = df[df['Has Online delivery'] == 'Yes']
top_online_delivery = online_delivery.sort_values(by='Aggregate rating', ascending=False).head(5)
print("üöö Top 5 Restaurants Offering Online Delivery:\n")
print(top_online_delivery[['Restaurant Name', 'City', 'Aggregate rating', 'Cuisines']])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter: Restaurants that offer online delivery
online_delivery = df[df['Has Online delivery'] == 'Yes']

# Sort by rating and select top 5
top_online_delivery = online_delivery.sort_values(by='Aggregate rating', ascending=False).head(5)

# Print data slice
print("üöö Top 5 Restaurants Offering Online Delivery:\n")
print(top_online_delivery[['Restaurant Name', 'City', 'Cuisines', 'Aggregate rating', 'Votes']])

# Visualize
plt.figure(figsize=(10, 6))
sns.barplot(x='Aggregate rating', y='Restaurant Name', data=top_online_delivery, palette='YlGnBu')
plt.title("üöö Top 5 Restaurants Offering Online Delivery")
plt.xlabel("Aggregate Rating")
plt.ylabel("Restaurant Name")
plt.xlim(4.0, 5.0)
plt.grid(axis='x', linestyle='--')
plt.show()
```

# Top 5 Budget-Friendly Restaurants (Price Range = 1 and Rating > 4.0)

```python
budget_best = df[(df['Price range'] == 1) & (df['Aggregate rating'] > 4.0)]
top_budget = budget_best.sort_values(by='Aggregate rating', ascending=False).head(5)
print("üí∏ Top 5 Budget-Friendly Restaurants:\n")
print(top_budget[['Restaurant Name', 'City', 'Aggregate rating', 'Cuisines', 'Votes']])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter: Price range = 1 and rating > 4.0
budget_friendly = df[(df['Price range'] == 1) & (df['Aggregate rating'] > 4.0)]

# Sort by rating and select top 5
top_budget = budget_friendly.sort_values(by='Aggregate rating', ascending=False).head(5)

# Show data
print("üí∏ Top 5 Budget-Friendly Restaurants:\n")
print(top_budget[['Restaurant Name', 'City', 'Cuisines', 'Aggregate rating', 'Votes']])

# Visualization
plt.figure(figsize=(10, 6))
sns.barplot(x='Aggregate rating', y='Restaurant Name', data=top_budget, palette='Purples')
plt.title("üí∏ Top 5 Budget-Friendly Restaurants")
plt.xlabel("Aggregate Rating")
plt.ylabel("Restaurant Name")
plt.xlim(4.0, 5.0)
plt.grid(axis='x', linestyle='--')
plt.show()
```
